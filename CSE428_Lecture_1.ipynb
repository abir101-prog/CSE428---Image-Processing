```python
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = (2, 2)
plt.rcParams['figure.dpi'] = 200  # pixel per inch
```

***List vs Numpy Array***

```python
x = [1, 2, 3] # list
y = np.array([1, 2, 3])  # numpy array
print(y+10) 
# numpy operation works elementwise
z = np.array([[1, 2], 
              [3, 4]])
print(z[1][0])  #[r][c]
# z[a:b] == from a to b-1
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
print(x[6:3:-1])
# sliced a reversed array
```

```python
print(x[:10])
# if index exceeds, it will simply go to the max
```

```python
w = np.array([[1, 2 ,3 , 4],
             [4, 2, 3, 4],
             [4, 6, 7, 1],
             [2, 4 ,6, 1]])
print(w)
```

```python
print(w[2:4, 1:3])
```

***Image Input Output***

```python
img = io.imread('femto.jpg')
```

```python
print(img.shape) #(H, W), GrayScale image, only 1 channel
```

```python
io.imshow(img)
plt.axis('off')  # not to show the value on axis
```

```python
me = io.imread('Me.jpg')
print(me.shape)  # since it is RGB image, there are 3 channels
```

```python
io.imshow(me)
plt.axis('off')
```

```python
# To make subplots:
plt.rcParams['figure.figsize'] = (10, 10)   # make figure bigger

plt.subplot(1, 3, 1)  # (R, C, where to place current plot)
io.imshow(me[:, :, 0])
plt.title('red')
plt.axis('off')

plt.subplot(1, 3, 2)
io.imshow(me[:, :, 1])
plt.title('green')
plt.axis('off')

plt.subplot(1, 3, 3)
io.imshow(me[:, :, 2])
plt.title('blue')
plt.axis('off')
```

```python
# now about png...
buu = io.imread('buu.PNG')
print(buu.shape)  # there is an extra channel for apacity: often called alpha channel
```

**Remarks**:
1. jpg image has 3channels, png has 4, 4th is called alpha channel
2. In python index (r, c): first row then column
3. For analysing contast of an image, least color channel is preferable
4. slice (a:b:h) , h=step size
5. functions of skimage.io: imread, imshow
